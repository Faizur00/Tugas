```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

print("setup completed")
```

```python
house_data = pd.read_csv("melb_data.csv")

print(house_data.columns)
```

```python
# choosing the prediction target
y = house_data.Price
print(y)
```

```python
# chooosing the features
data_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = house_data[data_features]
print(x)
```

```python
# inspecting our features
print(x.describe())
print("\n \n \n")
print(x.head())
```

```python
# Building a model based on these data
# step to buildin an ML model:
# -Define: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
# -Fit: Capture patterns from provided data. This is the heart of modeling.
# -Predict: Just what it sounds like
# -Evaluate: Determine how accurate the model's predictions are.
```

```python
# Model define(we define teh model with random at this case)
data_model = DecisionTreeRegressor(random_state=1)

# then we fit the model
data_model.fit(x,y)
```

```python
# Making prediction based on our model
print(x.head())
print("\nthe model prediction are:\n")
prediction = data_model.predict(x.head())
print(prediction)
```
